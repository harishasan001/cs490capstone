```python
import pandas as pd
```

```python
df1 = pd.read_csv("NYC_Air_Quality.csv")
df2 = pd.read_csv("NYC_Crime.csv")
df3 = pd.read_csv("NYC_Property_Sales.csv")
```

```python
print("Columns in NYC_Air_Quality.csv:", df1.columns)
print("\nColumns in NYC_Crime.csv:", df2.columns)
print("\nColumns in NYC_Property_Sales.csv:", df3.columns)
```

```python
print("Rows in NYC_Air_Quality.csv:", len(df1))
print("\nRows in NYC_Crime.csv:", len(df2))
print("\nRows in NYC_Property_Sales.csv:", len(df3))
```

```python

df1_years = df1['Start_Date'].str[:4].unique()  
years_in_df1 = len(df1_years)

df2['year'] = pd.to_datetime(df2['arrest_date']).dt.year  
years_in_df2 = len(df2['year'].unique())

df3['year'] = pd.to_datetime(df3['SALE DATE']).dt.year  
years_in_df3 = len(df3['year'].unique())

print(f"Years in NYC_Air_Quality data: {years_in_df1}")
print(f"Years in NYC_Crime data: {years_in_df2}")
print(f"Years in NYC_Property_Sales data: {years_in_df3}")
```
