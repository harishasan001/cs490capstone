```python


def get_high_correlations(df, threshold):
    # Compute correlation matrix
    correlation_matrix = df.corr()
    
    # Extract pairs above threshold
    high_corr_pairs = {}
    for col in correlation_matrix.columns:
        for index in correlation_matrix.index:
            if threshold < correlation_matrix.loc[index, col] < 1:  # 1 means perfect correlation, so we exclude that
                high_corr_pairs[(index, col)] = correlation_matrix.loc[index, col]
                
    return high_corr_pairs

# For df1
high_corr_pairs_df1_07 = get_high_correlations(df1, 0.7)
high_corr_pairs_df1_05 = get_high_correlations(df1, 0.5)

# For df2
high_corr_pairs_df2_07 = get_high_correlations(df2, 0.7)
high_corr_pairs_df2_05 = get_high_correlations(df2, 0.5)

# For df3
high_corr_pairs_df3_07 = get_high_correlations(df3, 0.7)
high_corr_pairs_df3_05 = get_high_correlations(df3, 0.5)



```

```python
# Print the results
print("For df1 - Pairs with correlation greater than 0.7:")
for pair, corr_value in high_corr_pairs_df1_07.items():
    print(pair, corr_value)

print("\nFor df1 - Pairs with correlation greater than 0.5:")
for pair, corr_value in high_corr_pairs_df1_05.items():
    print(pair, corr_value)
```

```python

# Convert 'SALE PRICE' to numeric
df3['SALE PRICE'] = pd.to_numeric(df3['SALE PRICE'], errors='coerce')

# Convert 'Data Value' in the air quality dataframe to numeric
df1['Data Value'] = pd.to_numeric(df1['Data Value'], errors='coerce')

# Median sale price by borough
median_sale_prices = df3.groupby('borough')['SALE PRICE'].median()

# For crime counts
crime_counts = df2.groupby('borough').size()

# For air quality
median_air_quality = df1.groupby('Borough')['Data Value'].median()

# Consistent color scheme
colors = plt.cm.viridis(np.linspace(0, 1, len(median_sale_prices)))

# Plotting the bar charts
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

def ensure_citywide_last(series):
    if 'Citywide' in series:
        citywide_val = series['Citywide']
        series = series.drop('Citywide')
        series['Citywide'] = citywide_val
    return series

median_air_quality = ensure_citywide_last(median_air_quality)
# Create colors for the air quality bars
air_quality_colors = [mcolors.to_rgba('grey') if borough == 'Citywide' else color 
                    for borough, color in zip(median_air_quality.index, colors)]



# Plot median sale price by borough
axs[0].bar(median_sale_prices.index, median_sale_prices.values, color=colors)
axs[0].set_title('Median Sale Price by Borough')
axs[0].set_ylabel('Sale Price')
axs[0].set_xlabel('Borough')

# Plot crime count by borough
axs[1].bar(crime_counts.index, crime_counts.values, color=colors)
axs[1].set_title('Crime Counts by Borough')
axs[1].set_ylabel('Number of Crimes')
axs[1].set_xlabel('Borough')

# Plot median air quality by borough with Citywide in grey
axs[2].bar(median_air_quality.index, median_air_quality.values, color=air_quality_colors)
axs[2].set_title('Median Air Quality by Borough')
axs[2].set_ylabel('Air Quality Metric')
axs[2].set_xlabel('Borough')

plt.tight_layout()
plt.show()
```
